# Store US Stock Prices in HDF5 Format

## Imports

```python
import nasdaqdatalink
nasdaqdatalink.read_key(filename="./.corporatenasdaqdatalinkapikey")
```

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
idx = pd.IndexSlice
```

## Load source data

```python
data = nasdaqdatalink.get_table('MER/F1', compnumber="39102", paginate=True)
```

```python
df = pd.read_csv("../data/us_stocks.csv",
                  parse_dates=['date'],
                  index_col=['ticker', 'date'])
```

```python
df.head()
```

```python
df.info(show_counts=True)
```

```python
df.describe()
```

## Apply required transformations

```python
df = (df.loc[idx[:, '2000':], :]
      .filter(like='adj')
      .rename(columns=lambda x: x.replace('adj_', ''))
      .dropna())
```

## Plot data

```python
df.head()
```

```python
symbol = 'AAPL'
```

```python
aapl_data_close = df.xs(symbol, level='ticker')['close']
aapl_data_volume = df.xs(symbol, level='ticker')['volume']

aapl_data_close.plot()
plt.title(f'{symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()

aapl_data_volume.plot()
plt.title(f'{symbol} Stock ')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, figsize=(14, 6), sharex=True)

s = df.loc[symbol, 'close']

s.plot(rot=0, ax=axes[0], title=f'{symbol} Close Price')

s.pct_change().plot(rot=0, ax=axes[1], title=f'{symbol} Daily Returns')

axes[1].set_xlabel('')

sns.despine()

fig.tight_layout()
```

## Store in HDF5 format

```python
df.to_hdf('../data/stock_prices.h5', 'us_stocks')
```
